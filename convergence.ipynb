```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
deltas = np.load('./data/npy/deltas.npy')
deltas = deltas[~np.isnan(deltas)]
```

```python
# app = [0.160479, 0.219777, 0.113650, 0.086468, 0.058224]
# deltas = np.append(deltas, app)
deltas = pd.DataFrame(deltas)
```

```python
my_dpi = 72
matplotlib.rcParams.update({'font.size': 16})
plt.figure(figsize=(1500/my_dpi, 800/my_dpi), dpi=my_dpi)
plt.title('Convergence: Plot shows diff in spot-to-cell probabilities between two successive iterations',  fontsize=20)
plt.ylabel('Delta', fontsize=18)
plt.xlabel('Iteration', fontsize=18)
plt.plot(deltas) # plotting by columns
plt.savefig('convergence.png', bbox_inches = 'tight')
plt.show()
```
